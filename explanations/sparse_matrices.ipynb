```python
# Sparce matrices, store the non 0 values
## the below demonstrates the exponential relationship between memory
## storing all the zeroes is wasteful, so store only the nonzero items
## '%matplotlib notebook' is call a magic command, it makes it interactive
## '%matplotlib inline' is another see the different when you run them.
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 1e6, 10)
plt.plot(x, 8.0 * (x**2) / 1e6, lw=5) 

plt.xlabel('size n')

plt.ylabel('memory [MB]')
```

# Explanation

Sparse matrices only stores non-zero values. This means on larger data sets you will save on memory.

## Example of a Compressed Sparse Row Matrix

```python
from scipy import sparse
eye = np.eye(5)
print("Numpy array with sparse :\n{}".format(eye))

sparse_matrix = sparse.csr_matrix(eye)
print('\nSparse matrix using a Compressed Sparse Row matrix (CSR): \n{}'.format(sparse_matrix))
```

As you can see, it removes thee unnecessary non-zero values. Imagine this but ~~millions, billions,~~ trillions of rows.

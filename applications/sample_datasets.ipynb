```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import mglearn
from IPython.display import display

```

# Example of a two class classification (binary)

The plot has the first feature on the xAxis and the second on the yAxis.

```python
#generate synthetic dataset
X, y = mglearn.datasets.make_forge()

#plot dataset

mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.legend(['Class 0', 'Class 1'], loc=4)
plt.xlabel('First feature')
plt.ylabel('Second feature')
print('X.shape: {}'.format(X.shape))
```

# Illustrating regression algorithms

This is using the synthetic wave dataset.

```python
X, y = mglearn.datasets.make_wave(n_samples=40)
plt.plot(X, y, 'o')
plt.ylim(-3, 3)
plt.xlabel('Feature')
plt.ylabel('Target')
```

Both datasets are done with low dimensional datasets because they are easily visualised.

If datasets have few features they are called ___low-dimensional datasets___

They are easily represented on paper.

___High-dimensional datasets___ might not work as well being represented in this way.
